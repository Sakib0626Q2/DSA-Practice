```python
'''
Implement pow(x, n), which calculates x raised to the power n (i.e., xn).
Example 1:

Input: x = 2.00000, n = 10
Output: 1024.00000
Example 2:

Input: x = 2.10000, n = 3
Output: 9.26100
Example 3:

Input: x = 2.00000, n = -2
Output: 0.25000
Explanation: 2-2 = 1/22 = 1/4 = 0.25

Constraints:
-100.0 < x < 100.0
-231 <= n <= 231-1
n is an integer.
Either x is not zero or n > 0.
-104 <= xn <= 104
'''
```

```python
def power(x, n):
    a = 1
    if n>0:
        for i in range(n):
            a *= x
    else:
        # abs(n)
        for i in range(-n):
            a *= x
        a = 1/a
    return float(a)
```

```python
print(power(4, -1))
print(power(2, 3))
print(power(3, 3))
```
